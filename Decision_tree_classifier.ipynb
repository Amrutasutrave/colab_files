**`DECISION TREE`**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report # Corrected import statement
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv('/content/WineQT.csv')
```

```python
#define feature and target column
X=df.iloc[:,:-1]#feature
Y=df.iloc[:,-2] #selecting second last col   #target
```

```python
Y.head()
```

```python
#features and target relation & outliers
for Column in X:
  plt.figure()
  sns.distplot(X[Column])
  plt.show()
  #decide skewness if +ve right side moving, -ve left side moving, normal if middle
  #decide skewness on describe function see its mean and its 50% value
```

```python
#visualizing target col
plt.figure()
sns.countplot(x=Y)
plt.show()
```

```python
#boxplot shows the all outliers on all col
for col in X:
  plt.figure()
  sns.boxplot(data=df,x='quality',y=col)
  plt.show()
```

```python
#checking coorelation
#multi colinearity
plt.figure()
sns.heatmap(X.corr(),annot=True)
plt.show()
```

```python
#train to split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
#Decision tree model
model=DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
#testing score
y_pred=model.predict(x_test)
```

```python
#testing model on testing data
print(classification_report(y_test, y_pred))
```
