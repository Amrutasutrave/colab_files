```python
#dropping null value col name 'Cuisines'
df = df.drop('Cuisines', axis=1)
```

```python
#shows the how many C & R are in dataset
df.shape
```

```python
#separating feature & target
X=df.iloc[:,:-1]#-------------F
Y=df.iloc[:,-1]#-------------T
```

```python
Y.head()
```

```python
#differntiating numerical & categorical column
n_cols=df.select_dtypes(include='number').columns
c_cols=df.select_dtypes(exclude='number').columns
```

```python
#Finding outliers
for col in n_cols:
  q1=df[col].quantile(0.25)
  q3=df[col].quantile(0.75)
  IQR=q3-q1
  lb=q1-1.5*IQR
  ub=q3+1.5*IQR
  outliers=df[(df[col]<lb) | (df[col]>ub)][col]
  print(f"{col}{outliers.tolist()}")
```

```python
#removing outliers
new_df = df.copy()
for col in df.select_dtypes(include='number'):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    new_df = new_df[(new_df[col] >= lower_bound) & (new_df[col] <= upper_bound)]
print(f"Original DataFrame size: {df.shape}")
print(f"New DataFrame size: {new_df.shape}")
```

```python
new_df.shape
```

```python
# Creating box plots for all numeric columns
plt.figure(figsize=(12, 8))
for i, col in enumerate(n_cols):
    plt.subplot(1, len(n_cols), i+1)
    sns.boxplot(y=new_df[col])
    plt.title(f' {col}')
    plt.ylabel('')
plt.tight_layout()
plt.show()
```
